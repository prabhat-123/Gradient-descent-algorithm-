# Gradient Descent Algorithm for making prediction

We are using gradient descent algorithm to train our model. The most important variables to be consider while doing gradient descent algorithm is how to choose for learning rates and for how many iterations a model has to learn to minimize the cost function. Here we have tried number of guesses to get the values for number of iterations and learning rates for our model in such a way that cost function be minimized. Finally, after trying several guesses, finally we have choosen the value for the learning rates and number of iterations to minimize the cost function. 

```python
def gradient_descent(x,y):
    m_updated = b_updated = 0
    iterations = 1000
    n = len(x)
    learning_rate = 0.00021
    costs = []
    ypredictor = []
    for i in range(iterations):
        y_predicted = m_updated * x + b_updated
        cost = (1/n) * sum([val**2 for val in (y-y_predicted)])
        md =  -(2/n) * sum(x*(y - y_predicted))
        bd = -(2/n) * sum(y- y_predicted)
        m_updated = m_updated - (learning_rate * md)
        b_updated = b_updated - (learning_rate * bd)
        costs.append(cost)
        
        print( "m {} , b {}, cost {} iteration {}".format(m_updated,
                                                   b_updated,cost,i))
        
    return [y_predicted.tolist(), costs, m_updated , b_updated]
   

result = gradient_descent(x,y)
ypredicted = result[0]
costvalue = result[1]
m = result[2]
b = result[3]
print(costvalue[-1])


```
